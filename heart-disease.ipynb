## MODELOS DE ML

```python
base = df.copy()

X = base.drop(columns='target')
y = df['target']

X.shape, y.shape
```

```python
base
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)

X
```

```python
X_teste.shape, y_teste.shape
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, accuracy_score, classification_report
```

```python
from sklearn.naive_bayes import GaussianNB

nvb = GaussianNB()
nvb.fit(X_treino, y_treino)
```

```python
y_pred = nvb.predict(X_teste)

y_pred
```

```python
accuracy_score(y_teste, y_pred)
```

```python
cm = confusion_matrix(y_teste, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(colorbar=False, cmap='Blues');
```

```python
print(classification_report(y_teste, y_pred))
```

## TUNING

```python
from sklearn.model_selection import GridSearchCV

# GaussianNB
param_grid_gnb = {
    'var_smoothing': [
        1e-12, 1e-11, 1e-10,
        1e-9, 1e-8, 1e-7,
        1e-6, 1e-5
    ]
}

# Logistic Regression
param_grid_lr = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear'],
    'class_weight': [None, 'balanced'],
    'max_iter': [1000]
}

# SVM
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto', 0.01, 0.001],
    'class_weight': [None, 'balanced']
}
```

```python
from sklearn.model_selection import GridSearchCV

models = {
    "GaussianNB": (
        GaussianNB(),
        {
            "var_smoothing": [1e-12, 1e-11, 1e-10, 1e-9, 1e-8]
        }
    ),

    "LogisticRegression": (
        LogisticRegression(),
        {
            "C": [0.001, 0.01, 0.1, 1, 10, 100],
            "penalty": ["l1", "l2"],
            "solver": ["liblinear"],
            "class_weight": [None, "balanced"],
            "max_iter": [1000]
        }
    ),

    "SVM": (
        SVC(),
        {
            "C": [0.1, 1, 10, 100],
            "kernel": ["linear", "rbf"],
            "gamma": ["scale", "auto", 0.01, 0.001],
            "class_weight": [None, "balanced"]
        }
    ),
}

results = []

for name, (model, params) in models.items():

    print(f"\nTreinando {name}...")

    grid = GridSearchCV(
        estimator=model,
        param_grid=params,
        cv=5,
        scoring="f1_macro",  # ou accuracy
        n_jobs=-1
    )

    grid.fit(X_treino, y_treino)

    results.append({
        "Modelo": name,
        "Melhor Score": grid.best_score_,
        "Melhores Parâmetros": grid.best_params_
    })

    print(f"Melhor Score: {grid.best_score_:.4f}")
    print(f"Melhores Parâmetros: {grid.best_params_}")

# Resultado final ordenado
results_df = pd.DataFrame(results)
results_df = results_df.sort_values(
    by="Melhor Score",
    ascending=False
)

print(results_df)
```

## CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score

clf = SVC(kernel='rbf', C=100, gamma='scale', class_weight=None)

scores = cross_val_score(clf, X, y, cv=5, scoring='accuracy',  n_jobs=-1)

print("Scores:", scores)
print("Média:", scores.mean())
print("Desvio padrão:", scores.std())

clf = RandomForestClassifier()
scores = cross_val_score(clf, X, y, cv=5, scoring='accuracy',  n_jobs=-1)

print("Scores:", scores)
print("Média:", scores.mean())
print("Desvio padrão:", scores.std())
```

```python
from sklearn.model_selection import cross_val_score, KFold

modelos = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'SVM': SVC(kernel='rbf', C=100, gamma='scale', class_weight=None),
    'Logistic Regression': LogisticRegression(max_iter=1000, penalty='l1', solver='liblinear', C=1, class_weight='balanced'),
    'Gaussian NB': GaussianNB()
}

resultados = {}

for nome, modelo in modelos.items():
    scores_modelo = []

    for i in range(30):
        kfold = KFold(n_splits=10, shuffle=True, random_state=i)
        score = cross_val_score(modelo, X, y, cv=kfold, scoring='accuracy', n_jobs=-1)
        scores_modelo.append(score.mean())

    resultados[nome] = scores_modelo

resultados
```

```python
df_resultados = pd.DataFrame(resultados)
df_resultados.describe()
```

```python
df_resultados.head()
```

## AVALIAÇÃO DE DESEMPENHO

```python
# Um dos testes que precisamos fazer é do ANOVA, mas antes ANOVA só funciona quando a distribuição é normal

from scipy.stats import shapiro

shapiro(resultados['Random Forest']), shapiro(resultados['SVM']), shapiro(resultados['Logistic Regression']), shapiro(resultados['Gaussian NB'])

# H0 -> dados em distribuição normal

# alpha = 0.05
# se p >= 0.05: não rejeita H0
# se p < 0.05: rejeita H0

# Neste caso,
# Todos os algoritmos estão em distribuição normal
```

```python
# TESTE DO ANOVA

# H0: Não há diferença estatística entre os grupos
# H1: há diferença estatística entre os grupos

from scipy.stats import f_oneway

_, p = f_oneway(resultados['Random Forest'], resultados['SVM'], resultados['Logistic Regression'], resultados['Gaussian NB'])

if p < 0.05:
    print("Rejeita H0: Há diferença estatística")
```

```python
# TESTE DE TUKEY

tukey = df_resultados.melt(var_name='model', value_name='accuracy')
tukey

from statsmodels.stats.multicomp import MultiComparison

compara = MultiComparison(tukey['accuracy'], tukey['model'])
teste_estatistico = compara.tukeyhsd()
teste_estatistico.plot_simultaneous();
```
